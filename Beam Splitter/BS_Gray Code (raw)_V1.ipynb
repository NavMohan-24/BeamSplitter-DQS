```python
import numpy as np

#Importing standard Qiskit libraries
from qiskit import *
from qiskit.tools.jupyter import *
from qiskit.visualization import *
from qiskit.providers import *
from qiskit.circuit import Parameter

import matplotlib.pyplot as plt
```

```python
from qiskit.circuit import Parameter

e1 = Parameter('$\epsilon_1$')
e2 = Parameter('$\epsilon_2$')
e3 = Parameter('$\epsilon_3$')
#__________________________________

e4 = Parameter('$\epsilon_4$')
e5 = Parameter('$\epsilon_5$')
```

```python
def stepwise_circ2(a,b,e,qc):
    
    arr=np.sort([a,b])
    
    qc.barrier()
    qc.cx(arr[0],arr[1])
    
    qc.barrier()
    qc.rz(e,arr[1])
    qc.barrier()
    
    qc.cx(arr[0],arr[1])
    qc.barrier()
    return(qc)

def stepwise_circ3(a,b,c,e,qc):
    
    arr=np.sort([a,b,c])
    
    qc.barrier()
    qc.cx(arr[0],arr[1])
    qc.cx(arr[1],arr[2])
    
    qc.barrier()
    qc.rz(e,arr[2])
    qc.barrier()

    qc.cx(arr[1],arr[2])
    qc.cx(arr[0],arr[1])
    qc.barrier()
    return(qc)

def stepwise_circ4(a,b,c,d,e,qc):
    
    arr=np.sort([a,b,c,d])
    
    qc.barrier()
    qc.cx(arr[0],arr[1])
    qc.cx(arr[1],arr[2])
    qc.cx(arr[2],arr[3])
    
    qc.barrier()
    qc.rz(e,arr[3])
    qc.barrier()

    qc.cx(arr[2],arr[3])
    qc.cx(arr[1],arr[2])
    qc.cx(arr[0],arr[1])
    qc.barrier()
    return(qc)
    
```

## $U_{1937}$

```python
qc1=QuantumCircuit(4)

qc1.h([1,3])
stepwise_circ2(1,3,e1,qc1)
qc1.h([1,3])

qc2=QuantumCircuit(4)

qc2.rx(np.pi/2,[1,3])
stepwise_circ4(0,1,2,3,e1,qc2)
qc2.rx(-1*np.pi/2,[1,3])
#_______________________________
qc3=QuantumCircuit(4)

qc3.rx(np.pi/2,[1,3])
stepwise_circ2(1,3,e2,qc3)
qc3.rx(-1*np.pi/2,[1,3])

qc4=QuantumCircuit(4)

qc4.h([1,3])
stepwise_circ4(0,1,2,3,e2,qc4)
qc4.h([1,3])
#_______________________________
qc5=QuantumCircuit(4)

qc5.h([1,3])
stepwise_circ3(1,2,3,-1*e3,qc5)
qc5.h([1,3])

qc6=QuantumCircuit(4)

qc6.rx(np.pi/2,[1,3])
stepwise_circ3(1,2,3,-1*e3,qc6)
qc6.rx(-1*np.pi/2,[1,3])

qc7=QuantumCircuit(4)

qc7.h([1,3])
stepwise_circ3(0,1,3,-1*e3,qc7)
qc7.h([1,3])

qc8=QuantumCircuit(4)

qc8.rx(np.pi/2,[1,3])
stepwise_circ3(0,1,3,-1*e3,qc8)
qc8.rx(-1*np.pi/2,[1,3])
#__________________________________

circ1 = QuantumCircuit(4, name='Circ-1937')

circ1.append(qc1,[0,1,2,3])
circ1.append(qc2,[0,1,2,3])
circ1.append(qc3,[0,1,2,3])
circ1.append(qc4,[0,1,2,3])
circ1.append(qc5,[0,1,2,3])
circ1.append(qc6,[0,1,2,3])
circ1.append(qc7,[0,1,2,3])
circ1.append(qc8,[0,1,2,3])

#circ1 = circ1.bind_parameters({e1: np.pi/16})
circ1.barrier()
circ1.draw('mpl')
```

```python
circ=circ1.decompose()
circ.draw('mpl')
```

```python
# Parameters

p1=np.pi*(2+np.sqrt(3))/8
p2=np.pi*(2-np.sqrt(3))/8
p3=np.pi/8
p4=np.pi*(np.sqrt(2)+np.sqrt(6))/16
p5=np.pi*(np.sqrt(2)-np.sqrt(6))/16
```

```python
def trotterize(qc,circ1,circ2,circ3,circ4,n):
    for i in range(n):
        qc.append(circ1,range(4))
        qc.append(circ2,range(4))
        qc.append(circ3,range(4))
        qc.append(circ4,range(4))
        qc = qc.bind_parameters({e1: -1*p1/n, e2: -1*p2/n, e3: -1*p3/n, e4: -1*p4/n, e5: -1*p5/n})
        qc.barrier()
    return qc
```

```python
a = []
counts_of_all_circuits = []
for i in range(1,20):
    qc = QuantumCircuit(4)
    qc.x([1,3])
    qc.barrier()
    qc = trotterize(qc,circ1,circ2,circ3,circ4,i)
    qc.measure_all()
    qasm = Aer.get_backend('qasm_simulator')
    counts = execute(qc, qasm, shots=1024).result().get_counts()
    counts_of_all_circuits.append(counts)
    print(counts,i)
```
