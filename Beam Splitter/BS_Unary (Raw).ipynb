## Coherent State

```python
alpha = 1
N = 1/np.sqrt((np.exp(-alpha**2)*(1**2+alpha**2+alpha**4/2)))
beta = np.exp(-alpha**2/2)
a = beta
b = beta*alpha
c = beta*alpha**2/np.sqrt(2)

vect = [0,0,0,N*a,0,N*b,N*c,0]
vect
```

```python
qc = QuantumCircuit(6)
qc.x([4,5])
qc.initialize(vect,[0,1,2])
qc.swap(0,2)
qc.barrier()
qc.append(circ1,range(6))
qc.append(circ2,range(6))
qc.append(circ3,range(6))
qc.append(circ4,range(6))

qc.measure_all()
qc = qc.bind_parameters({e1: np.pi/16, e2: np.sqrt(2)*np.pi/16, e3: np.sqrt(2)*np.pi/16, e4: 2*np.pi/16})
qc.draw('mpl')
```

```python
qasm = Aer.get_backend('qasm_simulator')
counts = execute(qc, qasm, shots=2048).result().get_counts()

print(f'\n\nCounts: {counts}\n\n')
plot_histogram(counts)
```
